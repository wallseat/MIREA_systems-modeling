Задание 1

```python
from numpy import around,mod
import math
from pprint import pprint

def RAND(a,b,m,x,N,A,B):
  l = []
  for i in range(N):
    x = (a*x + b)       
    x = math.fmod(x,m)  
    l.append(A+(B-A)*(x/m))
  return l
```

```python
TZ = RAND(39,39,1000,1,10**2,4,12)
TS = RAND(39,39,1000,1,10**2,2,8)
```

Задание 2

```python
import random 
random.seed(10)

def simulate(TZ, TS, size):

  # Входной поток
  inter_arrival_time = TZ
  # Время обработки заявок сервером
  service_time = TS
  # Массив для времени прихода программ
  arrival_time = [0 for _ in range(size)]
  # Изначальный
  arrival_time[0] = inter_arrival_time[0]

  for i in range(1,size):
    arrival_time[i] = inter_arrival_time[i]+arrival_time[i-1]
  
  ts_begin = [0 for i in range(size)]
  w_queue = [0 for i in range(size)]
  ts_end = [0 for i in range(size)]
  periods = [0 for i in range(size)]

  ts_begin[0] = arrival_time[0]
  ts_end[0] = service_time[0]
  periods[0] = service_time[0]

  for i in range(1,size):
    # Начало обработки 
    ts_begin[i] = max(arrival_time[i],ts_end[i-1])

    # Время ожидания в очереди   
    w_queue[i] = ts_begin[i]-arrival_time[i]

    # Время обработки
    ts_end[i] = ts_begin[i] + service_time[i]  

    # Время, потраченное на обработку
    periods[i] = ts_end[i] - arrival_time[i]
  no_customer_who_are_waiting = len(list(filter(lambda x:x>0,w_queue)))

  return arrival_time, w_queue, no_customer_who_are_waiting, size
```

```python
arrival_time, w_queue, no_customer_who_are_waiting, size = simulate(TZ, TS, 100)
```

```python
print(f"Времена прихода программ в вычислительную систему: ")
pprint(arrival_time)
```
